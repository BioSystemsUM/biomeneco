# minimal_completion method usage example

```python
# the directory /Users/josediogomoura/gap_filling_dl/tests/data/original_model has all the models to test the gap-filling, we want to iterate over all the directories and run the gap-filling for each directory. each directory/folder corresponds to a model and each directory has the following name structure: model_* where * is the number of each directory. they are 6
import os
from src.gap_filling_dl.biomeneco.gapfiller import GapFiller

reactions_removed = {
    'model_1': ['R00161_in', 'R02023_in', 'R01021_in'],
    'model_2': ['R01416_in'],
    'model_3': ['R00200_in', 'R00462_in'],
    'model_4': ['R00248_in', 'R02689_in'],
    'model_5': ['R01978_in'],
    'model_6': ['R03236_in', 'R00549_in', 'R04364_in', 'R01134_in', 'R01665_in', 'R03066_in']
}


base_path = '/Users/josediogomoura/gap_filling_dl/tests/data/original_model'


for dir in os.listdir(base_path):
    if dir.startswith('model'):
        print(dir)
        folder_path = os.path.join(base_path, dir)
        gap_filler = GapFiller.from_folder(folder_path, False, objective_function_id='Biomass_assembly_C3_in')
        gap_filler.run(False, write_to_file=True, removed_reactions=reactions_removed[dir])  # optimize = False para usar o minimal_completion
```
